## 🔍 **Aperçu du Projet**
Ce projet est une application de **webscraping** développée en **Python**, permettant l'extraction de données depuis **YouTube**. Il repose sur une architecture **conteneurisée** avec **Docker** et intègre plusieurs services :

- **Flask** : API web pour interagir avec les données
- **Celery** : Gestion des tâches asynchrones
- **PostgreSQL** : Stockage des données
- **Redis** : File d'attente pour les tâches
- **Prometheus & Grafana** : Surveillance des performances
- **pgAdmin** : Interface d'administration de la base de données

## 🏰 **Architecture des Services**
Le projet repose sur une **architecture Docker Compose** intégrant plusieurs services :

- 🌐 **Flask** : API web exposée sur le port `8000`
- 🗃️ **PostgreSQL** : Base de données pour stocker les informations collectées
- 🛠️ **Redis** : Gestion de la file d'attente des tâches Celery
- 🛠️ **Celery** : Worker pour exécuter les tâches en arrière-plan
- ⏳ **Celery-Beat** : Scheduler pour planifier les tâches
- 📊 **Prometheus** : Monitoring des performances (`9090`)
- 📈 **Grafana** : Interface de visualisation (`3000`)
- 🔧 **pgAdmin** : Interface de gestion PostgreSQL (`5050`)

Tous les services communiquent via un réseau Docker personnalisé (`app-network`).

## 🚀 **Installation & Déploiement**

## 🚀 Étapes d'Installation

### 1️⃣ Télécharger le Projet
- Copier le dossier du projet sur votre ordinateur
- Ouvrir le dossier du projet dans un éditeur de code

### 2️⃣ Vérifier les Fichiers Nécessaires
<div style="background-color: #1e3a29; color: white; padding: 15px; border-radius: 8px;">
  Assurez-vous que les fichiers suivants sont présents dans le répertoire racine :
  <ul>
    <li><code>Dockerfile</code></li>
    <li><code>docker-compose.yml</code></li>
    <li><code>requirements.txt</code></li>
    <li><code>app.py</code></li>
    <li><code>prometheus.yml</code></li>
    <li><code>servers.json</code></li>
    <li>Dossiers <code>templates/</code> et <code>static/</code></li>
  </ul>
</div>

### 3️⃣ Construire les Conteneurs
Exécutez la commande suivante pour construire les images Docker :
```bash
docker-compose build
```
<div style="background-color: #1e3a29; color: white; padding: 10px; border-radius: 5px; font-style: italic; margin: 10px 0;">
  Cela téléchargera les dépendances système (Chrome, Chromedriver, etc.) et Python spécifiées dans le <code>Dockerfile</code>.
</div>

### 4️⃣ Lancer les Services
Démarrez tous les services en mode détaché :
```bash
docker-compose up -d
```

- Ouvrez `http://localhost:8000` pour vérifier que l'application Flask est opérationnelle
- Consultez les logs si nécessaire :
  ```bash

  ```
  <div style="font-style: italic; color: #a0aec0; margin: 5px 0;">
    Remplacez <code>&lt;nom-du-service&gt;</code> par <code>flask</code>, <code>db</code>, <code>redis</code>, etc.

### 7️⃣ Arrêter les Services (Facultatif)
Pour arrêter les conteneurs :
```bash
docker-compose down
```
Pour supprimer également les volumes (données persistantes) :
```bash
docker-compose down -v
```
